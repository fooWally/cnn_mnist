```
import numpy as np
import tensorflow as tf
from tensorflow.keras import  losses
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import models
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
digits = load_digits()  # dict
print('digits.target.shape =', digits.target.shape) #(1797,)
print('digits.images.shape =', digits.images.shape) #(1797, 8, 8)

#-------------------------------------------------
# get test dataset
# Images shape (8,8) for input_shape 
X = digits.images ; y = digits.target
# Split data into train and test subsets
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, shuffle=True)
# Add a channels dimension: 
Xtrain = Xtrain[..., tf.newaxis].astype("float32")
Xtest  = Xtest[..., tf.newaxis].astype("float32")
test_dataset = tf.data.Dataset.from_tensor_slices((Xtest, ytest)).batch(64)
```

```
#-------------------------------------------------------------------
#   After finishing the training:
#   RuntimeError: You must compile your model before
#   training/testing. Use model.compile(optimizer, loss)   

# load model: load the model whose architecture is below
#model = models.load_model('mnist.h5') # not trained model
# display model.
#model.summary()

model = models.load_model('mnist.h5') # trained model
print('model.metrics_names =', model.metrics_names)
# display model.
model.summary()
```

```
# Warning says compile , so do it manually
#-----------------------------------------------
# Evaluate the model on the test data REQUIRES
# model.compile(...) as following
loss_fn = losses.SparseCategoricalCrossentropy()
model.compile(loss=loss_fn, optimizer='adam')
print("Evaluate on test data")
#print("test loss, test acc:", results)
for test_img, test_label in test_dataset:
    results = model.evaluate(Xtest, ytest)
    print('result =', results)

# Generate predictions (probabilities -- the output of the last layer)
# on new data using `predict`
print("Generate predictions for 3 samples")
# predictions.shape (3, 10) = (num_sample , num_classes)
# a row is one sample with 10 columns being the scores for 10 classes
# column idx of the highest score is the label predicted
predictions = model.predict(Xtest[:3])
print("predictions.shape:", predictions.shape)
##for ps in predictions:
##    label = np.argmax(ps)
##    print('label =', label)
answers = np.argmax(predictions, axis=1) # axis is the row
print('answers = ', answers)
pred_labels = model.predict_classes(Xtest[:3])
print('pred_labels =', pred_labels)
print('ytest[:3] =', ytest[:3])
```
